
## Tumor Identification Problem
<img src=https://raw.githubusercontent.com/basveeling/pcam/master/pcam.jpg>

In this project we'll build a machine learning model that identify turmors in small (96x96 pixel)samples of a high resolution image. We're using a Convolutional Neural network that has been pretrained to recognize a large set of images. We'll retrain this model using examples from this data-set:
https://github.com/basveeling/pcam
and replacing the last layers of the model with new versions that can learn our new task.

In general the process has two learning phases
1. Train the last layers of the neural network leaving the pre-trained parameters of the Convolutional Neural Network fixed
2. Un-freeze the Convolutional Layers and 'fine-tune' the pre-trained model

**However** This process an many tunable prarameters, in the ML world we call the hyper-parameters.

* Model Parameters
    * Which Pre-trained Model should we use
    * How big should our new-layers be
    * How much dropout should we use (a parameter that prevents overfitting but increases training time)
 
* Training Parameters
    * BatchSize
    * EarlyStopping
    * Maximum Epochs
    
* Hardware Parameters
    * CPU vs GPU
    * How Many of each
    
    
All of which effect both the time, cost, and performance of the model
